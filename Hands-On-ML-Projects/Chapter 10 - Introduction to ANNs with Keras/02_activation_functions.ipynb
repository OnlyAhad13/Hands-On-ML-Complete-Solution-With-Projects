```python
#ReLU

def relu(z):
    return np.maximum(0, z)

plt.plot(z, relu(z), label='relu', color='red')
plt.title('ReLU Function')
plt.grid(True)
plt.legend()
plt.show()
```

```python
#Leaky ReLU

alpha = 0.1
def leaky_relu(z):
    return np.where(z > 0, z, alpha*z)

plt.plot(z, leaky_relu(z), label='leaky_relu', color='purple')
plt.title('Leaky ReLU Function')
plt.grid(True)
plt.legend()
plt.show()
```

```python
#ELU

alpha = 1.0
def elu(z):
    return np.where(z > 0, z, alpha*(np.exp(z)-1))

plt.plot(z, elu(z), label='elu', color='orange')
plt.title('ELU Function')
plt.grid(True)
plt.legend()
plt.show()
```

```python
#SELU

scale = 1.050701
alpha = 1.673263

def selu(z):
    return scale * np.where(z > 0, z, alpha*(np.exp(z)-1))

plt.plot(z, selu(z), label='selu', color='brown')
plt.title('SELU Function')
plt.grid(True)
plt.legend()
plt.show()
```

```python
#Softplus

def softplus(z):
    return np.log1p(1 + np.exp(z))

plt.plot(z, softplus(z), label='softplus', color='pink')
plt.title('Softplus Function')
plt.grid(True)
plt.legend()
plt.show()
```
